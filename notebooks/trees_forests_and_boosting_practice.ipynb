All of decision trees, random forests, and gradient boosting models can be used in either regression problems or classification problems. You just have to use an appropriate loss function. In this notebook, I show very briefly how to use sklearn to train a decision tree, random forest, and gradient boosting model for regression and classification

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier, GradientBoostingRegressor, GradientBoostingClassifier
from sklearn.datasets import load_diabetes, load_breast_cancer
```

### Regression Practice

For the regression practice, we are going to use the diabetes dataset from scikit learn. Normally, I expect you to be more thoughtful with your feature selection and EDA, but for this practice notebook I just want to show how to train the different models, so I put no thought into the features.

```python
data = load_diabetes(as_frame = True)
#input variables
X = data['data'].values
#target variable
y = data['target'].values
```

```python
#decision tree
dt = DecisionTreeRegressor().fit(X, y)

plt.plot(dt.predict(X), y, 'k.')
plt.plot(dt.predict(X), dt.predict(X), 'b-')
plt.xlabel('Predictions')
plt.ylabel('Actual')
plt.title('Decision Tree Regressor on Diabetes Dataset')
plt.show()
```

Note that decision trees have a tendency to overfit, which explains why the predictions are so good. In practice, when you do machine learning you need to test your model on data that it wasn't trained on to check if it is overfitting or not. Feel free to experiment with different hyperparameters like the max depth, the min sample split, or any of the other ones we learned about in class!

```python
#random forest
rf = RandomForestRegressor().fit(X, y)

plt.plot(rf.predict(X), y, 'k.')
plt.plot(rf.predict(X), rf.predict(X), 'b-')
plt.xlabel('Predictions')
plt.ylabel('Actual')
plt.title('Random Forest Regressor on Diabetes Dataset')
plt.show()
```

Here we can see that the bagging of random forests has helped prevent us from overfitting! Again, feel free to expirement with the different random forest hyperparameters that we learned about in class, like all of the decision tree parameters or the number of trees in the forest.

```python
#gradient boost
gb = GradientBoostingRegressor().fit(X, y)

plt.plot(gb.predict(X), y, 'k.')
plt.plot(gb.predict(X), gb.predict(X), 'b-')
plt.xlabel('Predictions')
plt.ylabel('Actual')
plt.title('Gradient Boosting Regressor on Diabetes Dataset')
plt.show()
```

### Classification Practice

For the classification practice, we use the breast cancer dataset. Again, I am not doing any thoughtful feature selection or EDA here, I just want to show you how to implement the models.

```python
data = load_breast_cancer(as_frame = True)
#input variables
X = data['data'].values
#target variable
y = data['target'].values
```

```python
#decision tree
dt = DecisionTreeClassifier().fit(X, y)

print('Accuracy: ' + str(dt.score(X, y)))
```

```python
#random forest
rf = RandomForestClassifier().fit(X, y)

print('Accuracy: ' + str(rf.score(X, y)))
```

```python
#gradient boost
gb = GradientBoostingClassifier().fit(X, y)

print('Accuracy: ' + str(gb.score(X, y)))
```
